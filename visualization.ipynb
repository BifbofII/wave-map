```python
import numpy as np
import matplotlib.pyplot as plt
import plotly
import plotly.express as px
import plotly.graph_objects as go
import plotly.figure_factory as ff
```

```python
lat = np.load('data/visualization_example/lat.npy')
lon = np.load('data/visualization_example/lon.npy')
dat = np.load('data/visualization_example/wind_dat.npy')
```

```python
u_values = dat[:,:,0]
v_values = dat[:,:,1]
```

```python
y, x = np.meshgrid(lon,lat)
quiver_obj = plotly.figure_factory._quiver._Quiver(x, y, u_values, v_values, scale=.03, arrow_scale=0.3, angle=np.pi/9)
barb_x, barb_y = quiver_obj.get_barbs()
arrow_x, arrow_y = quiver_obj.get_quiver_arrows()


fig = go.Figure(data=go.Scattermapbox(lat=barb_x+arrow_x, lon=barb_y+arrow_y, mode='lines'),layout=None)
fig.update_layout(title='Wind direction', mapbox_style='open-street-map',
    mapbox=dict(zoom=4, center=dict(lat=x.mean(), lon=y.mean())),
    margin=dict(l=0, r=0, t=50, b=0))
fig.show()
```
